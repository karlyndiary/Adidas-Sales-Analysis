```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel(r'C:\Users\Karen Fernandes\anaconda3\Files\Projects\Adidas Sales Analysis\Adidas US Sales Datasets.xlsx', header = 4)
df.head()
```

```python
# Convert column headers to lowercase and make it snake case
df.columns = df.columns.str.lower().str.replace(' ', '_')
df.columns
```

```python
# dropping Unnamed: 0 column
df.drop('unnamed:_0', axis=1, inplace=True)
```

```python
df.columns
```

```python
# shape of the df
df.shape
```

```python
# checking for nulls and datatypes
df.info()
```

```python
# extract year, month, day columns from invoice_data
df['invoice_date'] = pd.to_datetime(df['invoice_date'])
df['year'] = df['invoice_date'].dt.year
df['month'] = df['invoice_date'].dt.month
df['day'] = df['invoice_date'].dt.day
```
